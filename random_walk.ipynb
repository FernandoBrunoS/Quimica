# Movimiento Browniano

El Movimiento Browniano es un fenómeno físico en el que partículas pequeñas (por ejemplo, moléculas) en un medio fluido se mueven de manera aleatoria. El movimiento Browniano fue descubierto por el botánico Robert Brown en 1827, quien observó el movimiento errático de partículas de polen en el agua. El movimiento Browniano se puede explicar utilizando la teoría cinética molecular, que establece que las partículas en un medio fluido se mueven de manera aleatoria debido a las colisiones con otras partículas en el medio.

Matemáticamente, el movimiento Browniano se puede describir utilizando una ecuación diferencial estocástica conocida como la ecuación de Langevin:

$$m \frac{d^2 X}{dt^2} = -\gamma \frac{dX}{dt} + \sqrt{2D} \frac{dW}{dt}$$

donde $X$ es la posición de la partícula, $m$ es su masa, $\gamma$ es el coeficiente de fricción del medio fluido, $D$ es el coeficiente de difusión (que está relacionado con la temperatura del medio) y $W$ es un proceso estocástico conocido como el movimiento Browniano.

La ecuación de Langevin establece que la aceleración de la partícula es igual a la fuerza de fricción (proporcional a la velocidad de la partícula) más una componente aleatoria debida al movimiento Browniano. Esta componente aleatoria se modela utilizando un proceso estocástico que sigue una distribución normal con media cero y varianza $dt$ (donde $dt$ es un pequeño incremento de tiempo).

# Distribución normal

La Distribución Normal, también conocida como la distribución de Gauss o campana de Gauss, es una distribución de probabilidad continua que se utiliza para modelar muchos fenómenos naturales. La distribución normal se caracteriza por su función de densidad de probabilidad (PDF) que tiene la forma:

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

donde $\mu$ es la media de la distribución y $\sigma$ es su desviación estándar. La PDF de la distribución normal es simétrica alrededor de la media $\mu$ y tiene su máximo en $\mu$. La integral de la PDF sobre todo el rango de valores es igual a uno, lo que significa que la probabilidad total de que la variable aleatoria tome cualquier valor es igual a uno.

La distribución normal se utiliza ampliamente en estadística para modelar muchos fenómenos naturales, como las alturas de la población, los errores de medición y las puntuaciones en pruebas estandarizadas. Además, la distribución normal es importante en la teoría del límite central, que establece que la distribución de la media de una muestra grande de datos se aproxima a una distribución normal a medida que aumenta el tamaño de la muestra.

# Histograma

Para crear un histograma, primero se debe tener un conjunto de datos continuos. Luego, se sigue el siguiente proceso:

Se divide el rango de valores de los datos en intervalos de igual tamaño, conocidos como "clases". La amplitud de cada clase se puede calcular como $amplitud = \frac{rango}{n}$, donde $rango$ es la diferencia entre el valor máximo y mínimo de los datos, y $n$ es el número de clases deseado.

Se cuenta la cantidad de datos que caen dentro de cada clase. Esta cantidad se conoce como la "frecuencia" de la clase.

Se dibuja un rectángulo para cada clase, cuya base es el intervalo de la clase y cuya altura es proporcional a la frecuencia de la clase. La altura se puede calcular dividiendo la frecuencia de la clase por el ancho de la clase.

Se repite el paso 3 para todas las clases, colocando los rectángulos uno al lado del otro.

Se agrega un título al histograma y etiquetas a los ejes X e Y, indicando qué variable se está representando y la frecuencia relativa de cada clase.

#Hacer que el caballo se mueva aleatoriamente en un tablero de ajedrez

*   Guardar en una lista de la que podamos elegir aleatoriamente
*   Elegir aleatoriamente

# ¿Qué es y en qué consiste el método de simulación Monte Carlo?
# GENERAL y con enfoque MOLECULAR

El método de simulación Monte Carlo es una técnica de simulación computacional que se utiliza para estimar la probabilidad de resultados en sistemas complejos. Este método se basa en la generación de números aleatorios para simular eventos aleatorios en un sistema. Aunque puede ser aplicado en una amplia variedad de campos, incluyendo finanzas, física, ingeniería y biología.

En el contexto de la simulación molecular, Monte Carlo es una técnica computacional que se utiliza para calcular propiedades termodinámicas de sistemas moleculares en equilibrio, tales como la energía libre de Gibbs, la entalpía y la entropía. El método se basa en la generación de configuraciones moleculares aleatorias, que son evaluadas mediante la aplicación de una función de energía potencial que describe las interacciones entre las moléculas del sistema.

El proceso de simulación Monte Carlo consiste en una serie de iteraciones en las que se generan configuraciones aleatorias del sistema, se evalúa la energía potencial de cada configuración y se calcula una probabilidad para aceptar o rechazar la configuración. La probabilidad de aceptación depende de la diferencia de energía entre la configuración actual y la nueva configuración propuesta, y puede ser ajustada mediante un parámetro llamado temperatura.

En general, cuanto más alta es la temperatura, mayor es la probabilidad de aceptar configuraciones con mayor energía, lo que permite explorar una mayor cantidad de conformaciones moleculares y aumentar la eficiencia de la simulación.

En resumen, el método de simulación Monte Carlo es una técnica de simulación computacional muy útil para calcular propiedades termodinámicas de sistemas complejos. En el contexto de la simulación molecular, el método se utiliza para generar configuraciones moleculares aleatorias y evaluar la energía potencial del sistema, lo que permite calcular propiedades termodinámicas como la energía libre de Gibbs, la entalpía y la entropía.

# Grafique un movimiento browniano con 1000 pasos en ($\mathbb{R}^{2}$)

```python
#Definimos n como el número de pasos
n = 1000

#Definimos varialbes pasos y los obtenemos con randn
pasosx =  np.random.randn(n)
pasosy =  np.random.randn(n)

#Creamos la variable "x" y "Y" y calculamos la suma de sus pasos
x = np.cumsum(pasosx)
y = np.cumsum(pasosy)

fig, ax = plt.subplots()

plt.plot(x,y,c='black')
plt.scatter(x,y,c=range(n),cmap='winter')

ax.axis('off')
plt.colorbar(label='paso',orientation='horizontal')
```

```python
# Configuración de los parámetros del movimiento Browniano
delta_t = 0.1 # Tamaño del paso en segundos
delta_x = 0.5 # Tamaño de cada paso en unidades de espacio
num_steps = 1000 # Número de pasos

# Generación de los cambios en la posición
dx = np.sqrt(delta_t) * delta_x * np.random.randn(num_steps)

# Cálculo de la posición del movimiento en cada paso
x = np.cumsum(dx)

# Graficación del movimiento
plt.plot(x)
plt.xlabel('Tiempo (pasos)')
plt.ylabel('Posición')
plt.title('Movimiento Browniano con 1000 pasos')
plt.show()
```
