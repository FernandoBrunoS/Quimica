#Aproximación del ángulo pequeño

Cuando $\theta \rightarrow θ$:

$$\sin \theta \approx \theta$$

$$\cos \theta \approx 1$$

$$\tan \theta \approx \theta$$

````python
import numpy as np # numerical python
import matplotlib.pyplot as plt # para graficar
````

## Graficar de $-\frac{\pi}{4}$ a $\frac{\pi}{4}$

````python
theta = np.linspace(-0.25*np.pi, 0.25*np.pi, 100) # Es un espacio lineal de números, igualmente espaciados, el tercer número dice cuántos puntos son

# Línea vertical en 0
plt.axvline(0,color='black',linestyle='dotted')
plt.axhline(0,color='black',linestyle='dotted')

# Gráficas
plt.plot(theta,np.sin(theta),label=r'$\sin \theta$',color='deepskyblue')
plt.plot(theta,np.cos(theta),label=r'$\cos \theta$',color='magenta')
plt.plot(theta,np.tan(theta),label=r'$\tan \theta$')

# Aproximaciones
plt.plot(theta,theta, label=r'$\theta$',color='deepskyblue',linestyle='dotted')
plt.plot(theta,np.ones((100)), label=r'$\theta$',color='magenta',linestyle='dashed')
#plt.axhline(1,color='magenta',linestyle='dooted')

# Nombre de los ejes
plt.xlabel(r'$\theta$', fontsize=20)
plt.ylabel(r'$f(\theta)$', fontsize=20)

# Agregar cuadrícula
plt.grid(linestyle='dotted',linewidth=0.5,color='lightgray')
plt.legend()
````

## Aproximación de Stirling

$$\ln n! \approx n\ln n - n$$

Ejemplos de factorial:

$$6!=1\cdot2\cdot3\cdot4\cdot5\cdot6=720$$

## Calcular factorial de un número

````python
num = 6

j = 1 # j es donde vamos a acumular las multiplicaciones
for i in range(1,num+1):
    j = j*i

print(j,np.math.factorial(num))
````

````python
import scipy.special as spe

n = np.arange(0.1,170,0.1)  # De 1 a 1001 cada 0.1

apr = n*np.log(n)-n
real = np.log(spe.factorial(n))


# Graficar

plt.plot(n,real,label='Real')
plt.plot(n,apr,label='Aproximacion')

plt.xscale("log")
plt.yscale("log")

plt.legend()
````

## Se lee el archivo

````python
import pandas as pd

mol = pd.read_csv('eter.xyz',skiprows=2, header=None,delim_whitespace=True)
mol.columns = ['at','x','y','z']
#mol
````

# Asignar color y tamaño a los átomos

````python
# Crear un diccionario de colores

color = {"O":"red","C":"gray","H":"white"}

size = {"O":0.73,"C":0.77,"H":0.37}
````

## Aplicar el diccionario a una columna de Pandas

````python
mol["color"] = mol['at'].map(color)
mol["size"] = mol['at'].map(size)
mol
````

$$d = \sqrt{(x_{2}-x_{1})^{2}+(y_{2}-y_{1})^{2}+(z_{2}-z_{1})^{2}}$$

$$d = \sqrt{\Delta x^{2}+\Delta y^{2}+\Delta z^{2}}$$

## Graficar en 3D

````python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib.animation import FuncAnimation,PillowWriter

from matplotlib import rc
rc('animation',html='jshtml')

fig = plt.figure()
ax = fig.add_subplot(projection='3d') # Si no, no es 3D

ax.scatter(mol['x'],mol['y'],mol['z'],c=mol["color"],s=100*mol["size"],edgecolor='black',alpha=1)



#### DIBUJAR ENLACES

for i in range(len(mol["at"])):
   x1,y1,z1 = mol["x"][i],mol["y"][i],mol["z"][i]
   for j in range(len(mol["at"])):
      x2,y2,z2 = mol["x"][j],mol["y"][j],mol["z"][j]
      d = ((x2-x1)**2+(y2-y1)**2+(z2-z1)**2)**0.5
      if d<1.6 and (i!=j):  # if 0<d<1.6: esta es otra manera
        ax.plot([x1,x2],[y1,y2],[z1,z2],color='gray')

ax.axis('off')

#ax.set_facecolor('k')

# Emcontrar el ancho
anchox = np.max(mol["x"])-np.min(mol["x"])
anchoy = np.max(mol["y"])-np.min(mol["y"])
anchoz = np.max(mol["z"])-np.min(mol["z"])

ANCHO = np.max([anchox,anchoy,anchoz])
cogx, cogy, cogz = np.mean(mol["x"]), np.mean(mol["y"]),np.mean(mol["z"])

# Centro geométrico

ax.scatter(cogx,cogy,cogz,color='yellow',s=1000)

#Ajustar los límites de los ejes

ax.set_xlim(cogx-0.5*ANCHO,cogx+0.5*ANCHO)
ax.set_ylim(cogy-0.5*ANCHO,cogy+0.5*ANCHO)
ax.set_zlim(cogz-0.5*ANCHO,cogz+0.5*ANCHO)

# Crear una función que lo rote

def rotacion(ang):
    ax.view_init(azim = ang)
    plt.draw()

anim = FuncAnimation(fig,rotacion,frames=range(0,360,10))
anim.save('corona.gif',fps=640)
anim
````

## Encontrar el centro geométrico de la molécula

````python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib.animation import FuncAnimation,PillowWriter

from matplotlib import rc
rc('animation',html='jshtml')

fig = plt.figure()
ax = fig.add_subplot(projection='3d')


a,b,c = 0,0,0

ax.scatter(a,b,c)

def esfera(ax,a,b,c,color,r):
    phi = np.linspace(0,2*np.pi,100)
    theta = np.linspace(0,np.pi,100)
    PHI,THETA = np.meshgrid(phi,theta)

    X = a + r*np.cos(PHI)*np.sin(THETA)
    Y = b + r*np.sin(PHI)*np.sin(THETA)
    Z = c + r*np.cos(THETA)

    ax.plot_surface(X,Y,Z, alpha=0.5,color=color)
    return ax

esfera(ax,1,2,3,'white',3)
esfera(ax,1,2,3,'red',1)
ax.axis('off')
````

$$x = R \cos \phi \sin\theta$$
$$x = R \sin \phi \sin\theta$$
$$x = R \cos \theta$$

$$\phi \in [0,2\pi]$$
$$\theta \in [0,\pi]$$
