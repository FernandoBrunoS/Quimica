# Historia de Pi

Pi es una constante matemática que representa la relación entre la circunferencia y el diámetro de un círculo. Su valor numérico es aproximadamente 3.14159 , aunque su representación decimal es infinita y no periódica. La búsqueda de la precisión en el cálculo de pi ha sido una preocupación matemática desde la antigüedad, siendo los antiguos egipcios y babilonios los primeros en aproximarse a su valor. En la antigua Grecia, el matemático Arquímedes obtuvo un método geométrico para calcular una aproximación más precisa de pi. En la Edad Media y el Renacimiento, matemáticos como Leonardo Fibonacci y Ludolph van Ceulen continuaron buscando formas de calcular pi con mayor precisión, y con el tiempo se desarrollaron algoritmos computacionales para calcular su valor con millas de millones de decimales.





# Algoritmo de metropolis

El algoritmo de Metropolis es un método de simulación utilizado en física, estadística y otras áreas de investigación que involucran la generación de muestras aleatorias de una distribución de probabilidad dada. El objetivo del algoritmo es generar muestras aleatorias de una distribución de probabilidad que sea difícil de muestrear directamente, utilizando una distribución de probabilidad de transición propuesta más fácil de muestrear. El algoritmo de Metropolis funcionará una secuencia de candidatos de muestras, y luego aceptará o rechazará cada candidato en función de una regla de aceptación que depende de la relación entre la distribución de probabilidad objetivo y la distribución de probabilidad propuesta. De esta manera, el algoritmo produce una secuencia de muestras que se aproxima a la distribución de probabilidad objetivo.

# - ¿Cómo se calcula la energía potencial de Coulomb entre dos partículas cargadas en unidades atómicas?

La energía potencial de Coulomb entre dos partículas cargadas en unidades atómicas se puede calcular usando la siguiente fórmula:

$U_{Culombio} = \frac{q_1 q_2}{4 \pi \epsilon_0 r_{12}} \hspace{1cm} (\frac{e^2}{4 \pi \epsilon_0 a_0})$

donde $q_1$ y $q_2$ son las cargas de las partículas, $r_{12}$ es la distancia entre las partículas y $\epsilon_0$ es la permitividad eléctrica en unidades atómicas. La constante $a_0$ es la radio de Bohr, que se utiliza para convertir las unidades atómicas a unidades SI.

# Modelo de Ising

Es un modelo *sencillo* de materiales que sufren un cambio brusco en sus propiedades magnéticas con la temperatura.

A T bajas ❄ → Ferromagnéticos

A T altas 🔥 → Paramagnéticos

- Es una red cuadrada bidimensional $N\times N$ o $N^{2}$
- Cada nudo tiene un espín $s(i,j)$ que puede ser 1 o -1, con igual probabilidad

- La energía de interacción del sistema es:

$$E(S)=-\frac{1}{2}\sum^{N}_{i=1}\sum^{N}_{j=1} s(i,j)[s(i,j+1)+s(i,j-1)+s(i+1,j)+s(i-1,j)]$$

- Condiciones periódicas de frontera

A la izquierda:
     $$s(-1,j) = s(N-1,j)$$

A la derecha:
     $$s(N,j) = s(0,j)$$

Arriba:
     $$s(i,N)=s(i,0)$$

Abajo:
     $$s(i,-1)=s(i,N-1)$$

- La probabilidad está dada por:
$$P(S)=\frac{1}{Z}e^{-\beta E(S) }$$
$$P(S)=\frac{1}{Z}e^{-\frac{E(S)}{k_{B}T} }$$

## Criterio de metropolis

1. Para el modelo de Ising $T\in[0,5]$ en un unidades atómicas

2. Se genera una condición inicial de espínes (ordenada o desordenada)

3. Elegir un espín al azar

4. Calcula el $\Delta E$ del intercambio aplicando las condiciones periódicas

$$\Delta E = 2 s(i,j)\big[s(i+1,j)+s(i-1,j)+s(i,j+1)+s(i,j-1)\big]$$

5. Si $\Delta E$ es negativa, se acepta el cambio.

6. Si no, se calcula la probabilidad y se lanza un número aleatorio.

```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation,PillowWriter

from matplotlib import rc
rc('animation',html='jshtml')

def dEnergia(m,i,j):
   aqui = m[i][j]
   # derecha
   if (i+1)<N:
      der = m[i+1][j]
   else:
      der = m[0][j]
  #izquierda
   if (i-1)>0:
      izq = m[i-1][j]
   else:
      izq = m[N-1][j]

   # arriba
   if (j+1)<N:
      arriba = m[i][j+1]
   else:
      arriba = m[i][0]

   #abajo
   if (j-1)>0:
      abajo = m[i][j-1]
   else:
      abajo = m[i][N-1]
   dE = 2*aqui*(der+izq+arriba+abajo)
   return dE



fig,ax = plt.subplots(figsize=(2,2))
ax.axis('off')

N = 16
m = np.random.choice([-1,1],size=(N,N))
#m = np.ones((N,N))

plt.imshow(m,cmap='bwr')

T = 10000

npasos = 10000
G=[]
conteo=0

for paso in range(npasos):
   i = np.random.randint(N)
   j = np.random.randint(N)
   dE = dEnergia(m,i,j)

   if dE<0: # se acepta el cambio
      m[i][j] *= -1
   elif T!=0 and ((1/N**2)*np.exp(-dE/T) > np.random.rand()): # lanza un num aleatorio
      m[i][j] *= -1
   #plt.imshow(m,cmap='bwr')
   if paso%10==0: #Si paso es multiplo de 100
      G.append(m.copy())
   if paso%1000==0:
      conteo += 1
      print(conteo*'●'+ (10-conteo)*'◌')

def animacion(i):
  ax.clear()
  ax.axis('off')
  plt.imshow(G[i],cmap='bwr',vmin=1 , vmax=1)
  plt.draw()




anim = FuncAnimation(fig,animacion,frames=range(len(G)))
#anim.save('ising.gif')
anim

```

## Actividad 3: Graficar el estado inicial y el estado final para 256,000 pasos para las temperaturas 0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5 y 5.0

```python
import numpy as np
import matplotlib.pyplot as plt

def ModeloIsing(T: float, npasos: int, N: int = 16):
    def dEnergia(m, i, j):
        espin_actual = m[i, j]
        vecinos = m[(i+1)%N, j] + m[(i-1)%N, j] + m[i, (j+1)%N] + m[i, (j-1)%N]
        return 2 * espin_actual * vecinos

    intervalo_progreso = max(1, npasos // 10)
    m = np.random.choice([-1, 1], size=(N, N))
    conteo = 0

    fig, ax = plt.subplots(1, 2, figsize=(8, 4))
    fig.suptitle(f'Modelo de Ising (T = {T}, N = {N})')

    ax[0].imshow(m, cmap='bwr', vmin=-1, vmax=1)
    ax[0].set_title('Estado Inicial')
    ax[0].axis('off')

    print(f"Iniciando simulación a T={T}...")
    for paso in range(npasos):
        i, j = np.random.randint(0, N, size=2)
        dE = dEnergia(m, i, j)

        # --- Lógica corregida para T = 0 ---
        if dE <= 0:
            # Siempre se acepta si la energía baja o es igual
            m[i, j] *= -1
        elif T > 0:
            # Solo si T > 0 calculamos la probabilidad de Boltzmann
            if np.random.rand() < np.exp(-dE / T):
                m[i, j] *= -1
        # Si T = 0 y dE > 0, no se hace nada (el espín no cambia)
        # ----------------------------------

        if (paso + 1) % intervalo_progreso == 0:
            conteo += 1
            print(f"Progreso: {conteo * '●'}{(10 - conteo) * '◌'} {conteo * 10}%")

    ax[1].imshow(m, cmap='bwr', vmin=-1, vmax=1)
    ax[1].set_title('Estado Final')
    ax[1].axis('off')

    plt.tight_layout()
    plt.show()
```

## 5.1 analizar la diferencia entre empezar con espines aleatorios y con todos iguales (aprox un parrafo)

El análisis de las condiciones iniciales en el modelo de Ising revela cómo el sistema alcanza el equilibrio térmico dependiendo de la energía disponible (temperatura). Al iniciar con espines aleatorios, el sistema posee una entropía inicial alta y debe "decidir" hacia qué estado de orden evolucionar, formando múltiples dominios o "islas" de espines alineados que compiten entre sí. Por el contrario, al empezar con todos los espines iguales (por ejemplo, todos en 1), el sistema parte de un estado de energía mínima y orden total. A temperaturas bajas (T=1.5), este orden es sumamente estable y el sistema permanece prácticamente inalterado porque no hay suficiente energía térmica para voltear los espines contra sus vecinos. Sin embargo, a temperaturas altas (T=5.0), la agitación térmica destruye rápidamente este orden inicial, demostrando que, sin importar el punto de partida, el sistema tenderá a un estado de desorden si la temperatura es lo suficientemente elevada, de forma similar a cómo una constante de fuerza k baja permite mayor libertad de movimiento en un oscilador

## Un paso de Monte Carlo es de $N^{2}$ intentos de cambio, en el que en teoría, todos los spines hayan podido cambiar.

```python
import numpy as np
import matplotlib.pyplot as plt

def SimularIsing_MCS(T, mcs, N=16):
    m = np.ones((N, N)) # Todos los espines en 1
    npasos = mcs * (N**2)

    for paso in range(npasos):
        i, j = np.random.randint(0, N, size=2)
        # Cálculo de dE con fronteras periódicas
        vecinos = m[(i+1)%N, j] + m[(i-1)%N, j] + m[i, (j+1)%N] + m[i, (j-1)%N]
        dE = 2 * m[i, j] * vecinos

        # Criterio de Metropolis
        if dE <= 0 or (T > 0 and np.random.rand() < np.exp(-dE / T)):
            m[i, j] *= -1

    # Cálculo de Magnetización Final
    mag_final = np.sum(m) / (N**2)
    return m, mag_final

# Ejecución y Análisis
temps = [1.5, 2.0, 5.0]
for t in temps:
    matriz_final, mag = SimularIsing_MCS(t, 1000) # 1000 MCS
    print(f"Temperatura: {t} | Magnetización final: {mag:.4f}")
```
